```python
from indicnlp.tokenize import indic_tokenize
def tokenization(indic_string):
    tokens = []
    for t in indic_tokenize.trivial_tokenize(indic_string):
        tokens.append(t)
    return tokens
df['Unnamed: 2'] = df['Unnamed: 2'].apply(lambda x: tokenization(x))
```
